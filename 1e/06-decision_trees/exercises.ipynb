```python
%%time
for i in range(X.shape[0]):
    if y[i] == 0:
        plt.plot(X[i], "xb")
    else:
        plt.plot(X[i], "^r")
```

```python
%%time
for i in range(X.shape[0]):
    if y[i] == 0:
        plt.plot(*X[i], "xb")
    else:
        plt.plot(*X[i], "^r")
```
