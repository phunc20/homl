```python
def easy_gray(rgb_image):
    gray_image = np.mean(rgb_image, axis=2)
    return gray_image
```

```python
gray_flower = easy_gray(flower)
gray_flower.shape
```

```python
plt.imshow(gray_flower, cmap="gray");
```
